```python
import matplotlib.pyplot as plt;
import numpy as np;
import scipy.optimize as opt;
```

 the "*.txt" file contains the "mass($1)" and "luminosity($2)" of all the stars in the ZAMS for Fe/H=-1

```python
data = np.loadtxt("M-L-z014.txt")
data2 = np.loadtxt("Mlow-L-z014.txt")
data3 = np.loadtxt("Mup-L-z014.txt")
data=np.transpose(data)
print(np.shape(data))
data.sort(axis=1)
data2=np.transpose(data2)
data3=np.transpose(data3)
print(np.shape(data2))
print(np.shape(data3))
```

I changed the LogL in MIST data to the "Luminosity(L_sun)" 

```python
Mdata = data[0]
Ldata = data[1]
print(Mdata)
print(Ldata)
```

```python
Mlowdata = data2[0] #### M <= 1.09 M_sun   (Eggleton et.al 1989)
Llowdata = data2[1] 
Mupdata = data3[0]  #### M >= 1.09 M_sun
Lupdata = data3[1]
print(Mlowdata)
print(Llowdata)
print(Mupdata)
print(Lupdata)
```

```python
plt.plot(Mdata, Ldata, ".", label=data)
plt.xlabel('M(M_s)',fontsize=16)
plt.ylabel('L(L_s)',fontsize=16)
plt.title('Z=0.014',fontsize=16)
```

```python
plt.loglog(Mdata,Ldata, ".")
plt.xlabel('LogM(M_s)',fontsize=16)
plt.ylabel('LogL(L_s)',fontsize=16)
plt.title('Z=0.014',fontsize=16)
```

```python
def funcL(M, alpha, beta, landa, delta, epsilon, zeta, nu ):
     return (alpha*M**5.5 + beta*M**11.0)/(landa + M**3.0 + delta*M**5.0 + epsilon*M**7.0 + zeta*M**8.0 +nu*M**9.5)
```

Using the tout et.al(1996) form of function.


```python
optimizedParameters,pcov = opt.curve_fit(funcL, Mdata, Ldata,maxfev=1000000000);
print(optimizedParameters);
```

```python
plt.plot(Mdata, funcL(Mdata, *optimizedParameters), label="fit");
plt.plot(Mdata, Ldata, ".", label="data")
plt.xlabel('M(M_s)',fontsize=16)
plt.ylabel('L(L_s)',fontsize=16)
```

the best fitted form of Tout's function to our data. I  checked it using Mathlab too.   
my first question: how did they find this form of function? as I realized their function is case based and it seems not a good fit to our data. 

```python
plt.plot(Mdata, funcL(Mdata, *optimizedParameters), label="fit");
plt.plot(Mdata, Ldata, ".", label="data")
plt.axes().set(xlim=(-2.5,15),ylim=(-100,1000))
```

```python
plt.loglog(Mdata, funcL(Mdata, *optimizedParameters), label="fit");
plt.loglog(Mdata, Ldata, ".", label="data")
```

I changed the form of function to a Quadratic  polynomial function.


```python
def func2L(M, a, b, c, d, e):
     return (a*M**4 + b*M**3+c* M**2.0 + d*M + e )
```

```python
optimizedParameters,pcov = opt.curve_fit(func2L, Mdata, Ldata,maxfev=100000);
print(optimizedParameters);
```

```python
plt.plot(Mdata, Ldata, ".", label="Data")
plt.plot(Mdata, func2L(Mdata, *optimizedParameters), label="fit")
```

```python
plt.plot(Mdata, Ldata, ".", label="Data")
plt.plot(Mdata, func2L(Mdata, *optimizedParameters), label="fit")
plt.axes().set(xlim=(-2.5,8),ylim=(-1000,1000))
```
