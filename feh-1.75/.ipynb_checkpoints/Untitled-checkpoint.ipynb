```python
def func(M, a, b, c):
    return a * np.exp(-b * M) + c
```

```python
optimizedParameters,pcov = opt.curve_fit(func, M, L,maxfev=100000);
print(optimizedParameters);
```

```python
plt.plot(M, L, ".", label="Data")
plt.plot(M, func(M, *optimizedParameters), label="fit")
```

```python
def func(Mup, a, b, c):
    return a * Mup**b+c
optimizedParameters,pcov = opt.curve_fit(func, Mup, Lup,maxfev=100000);
print(optimizedParameters);
```

```python
plt.plot(Mup, Lup, ".", label="Data")
plt.plot(Mup, func(Mup, *optimizedParameters), label="fit")
```

```python
#print(func(Mup, *optimizedParameters))
fitL=func(Mup, *optimizedParameters)
err=Lup-fitL
err=np.abs(err)
plt.plot(Mup,err,'')
plt.xlabel('M')
plt.ylabel('L-L_fit')
plt.title('errors')
```

```python
def func2L(M, c, e, f, q, k, l , p):
     return ( e*(M/c)**3.5+ f*(M)**q + k*(M/l)**2.5 )
```

```python
optimizedParameters,pcov = opt.curve_fit(func2L, M, L,maxfev=100000);
print(optimizedParameters);
```

```python
plt.subplot(121)
plt.plot(M, L, ".", label="Data")
plt.plot(M, func2L(M, *optimizedParameters), label="fit")
plt.subplot(122)
plt.axis([0,10,-2000,2000])
plt.plot(M, func2L(M,*optimizedParameters))
plt.plot(M, L, ".", label="Data")
```
